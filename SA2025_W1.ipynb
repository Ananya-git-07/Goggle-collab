```python
#your code here
weight = data['weight'].astype(float)
mask = (horsepower > 100) & (weight < 3000)
filtered_origins = data['origin'][mask]

# Count unique values
unique, counts = np.unique(filtered_origins, return_counts=True)
most_common_origin = unique[np.argmax(counts)]
print(most_common_origin)
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load your dataset
df = pd.read_csv('Cars.csv')
median_mpg = df['mpg'].median()
above_median = df[df['mpg'] > median_mpg].copy()

above_median['hp_to_weight'] = above_median['horsepower'] / above_median['weight']
max_ratio = above_median['hp_to_weight'].max()

best_cars = above_median[above_median['hp_to_weight'] == max_ratio]
print(best_cars[['name', 'hp_to_weight']])
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x='model_year', y='mpg', hue='origin', estimator='mean', ci=None)

plt.title('Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.show()
```

### Create a Seaborn scatterplot (or PairGrid) where:

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
# Group by name and count distinct years
grouped = df.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    mpg_std=('mpg', 'std'),
    mpg_avg=('mpg', 'mean')
).reset_index()

# Filter consistent cars
consistent = grouped[(grouped['appearances'] > 1) & (grouped['mpg_std'] < 1.0)]

# Sort
consistent_sorted = consistent.sort_values(by=['appearances', 'mpg_avg'], ascending=[False, False])

print(consistent_sorted[['name', 'appearances', 'mpg_avg']])
```
